```python
import tkinter as tk
from tkinter import messagebox, simpledialog

# Main task list
tasks = []

# Add a task
def add_task():
    task = task_entry.get().strip()
    if task:
        tasks.append(task)
        update_listbox()
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Empty Task", "Please enter a task before adding.")

# Delete selected task
def delete_task():
    selected = task_listbox.curselection()
    if selected:
        index = selected[0]
        task = tasks.pop(index)
        update_listbox()
        messagebox.showinfo("Task Deleted", f"Task '{task}' deleted.")
    else:
        messagebox.showwarning("No Selection", "Please select a task to delete.")

# Update selected task
def update_task():
    selected = task_listbox.curselection()
    if selected:
        index = selected[0]
        old_task = tasks[index]
        new_task = simpledialog.askstring("Update Task", f"Edit the task:", initialvalue=old_task)
        if new_task:
            tasks[index] = new_task.strip()
            update_listbox()
        else:
            messagebox.showwarning("Empty Task", "Updated task cannot be empty.")
    else:
        messagebox.showwarning("No Selection", "Please select a task to update.")

# Refresh the listbox
def update_listbox():
    task_listbox.delete(0, tk.END)
    for i, task in enumerate(tasks):
        task_listbox.insert(tk.END, f"{i + 1}. {task}")

# Create main window
root = tk.Tk()
root.title("To-Do List App")
root.geometry("500x400")
root.resizable(False, False)

# Widgets
title_label = tk.Label(root, text="üìù To-Do List", font=("Helvetica", 18, "bold"))
title_label.pack(pady=10)

# Entry field
task_entry = tk.Entry(root, font=("Helvetica", 14), width=30)
task_entry.pack(pady=5)

# Button frame
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

add_button = tk.Button(button_frame, text="Add Task", width=12, command=add_task)
add_button.grid(row=0, column=0, padx=5)

update_button = tk.Button(button_frame, text="Update Task", width=12, command=update_task)
update_button.grid(row=0, column=1, padx=5)

delete_button = tk.Button(button_frame, text="Delete Task", width=12, command=delete_task)
delete_button.grid(row=0, column=2, padx=5)

# Task listbox
task_listbox = tk.Listbox(root, font=("Helvetica", 12), width=50, height=10, selectbackground="skyblue")
task_listbox.pack(pady=10)

# Run the app
root.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox, simpledialog

# List to store tasks
tasks = []

# Function to add a new task
def add_task():
    task = entry.get()
    if task.strip() != "":
        tasks.append(task)
        update_listbox()
        entry.delete(0, END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

# Function to delete a selected task
def delete_task():
    try:
        selected_index = listbox.curselection()[0]
        tasks.pop(selected_index)
        update_listbox()
    except IndexError:
        messagebox.showwarning("Warning", "Please select a task to delete.")

# Function to update a selected task
def update_task():
    try:
        selected_index = listbox.curselection()[0]
        old_task = tasks[selected_index]
        new_task = simpledialog.askstring("Update Task", "Edit your task:", initialvalue=old_task)
        if new_task and new_task.strip() != "":
            tasks[selected_index] = new_task
            update_listbox()
        else:
            messagebox.showwarning("Warning", "Task cannot be empty.")
    except IndexError:
        messagebox.showwarning("Warning", "Please select a task to update.")

# Function to refresh the listbox
def update_listbox():
    listbox.delete(0, END)
    for task in tasks:
        listbox.insert(END, task)

# Main window
root = Tk()
root.title("To-Do List")
root.geometry("400x350")

# Title label
label = Label(root, text="To-Do List", font=("Arial", 16))
label.pack(pady=10)

# Entry for new task
entry = Entry(root, width=30, font=("Arial", 12))
entry.pack(pady=5)

# Buttons
add_btn = Button(root, text="Add Task", width=10, command=add_task)
add_btn.pack(pady=5)

update_btn = Button(root, text="Update Task", width=10, command=update_task)
update_btn.pack(pady=5)

delete_btn = Button(root, text="Delete Task", width=10, command=delete_task)
delete_btn.pack(pady=5)

# Listbox to show tasks
listbox = Listbox(root, width=40, height=10, font=("Arial", 12))
listbox.pack(pady=10)

root.mainloop()
```
